```python
files=glob.glob('C:\\Users\\Steph\\OneDrive\\Documents\\Sandra Hospital analysis\\2020-2024 Analysis\\Pharmacy/*.csv')

#Create empty datafram to place values
fullpharm = pd.DataFrame()

#Loop through each file and populate the dataframe
for file in files:
    df=pd.read_csv(file)
    fullpharm = pd.concat([fullpharm, df], ignore_index = True)
    
fullpharm.head(3)
```

```python
#Drop unneccesary colulms
fullpharm.drop(fullpharm.columns[[0,3,4,5,7,10]], axis=1, inplace=True)
```

```python
#drop duplicate headings
fullpharm.drop(fullpharm[fullpharm['Date'] == "Date"].index, inplace = True)
```

To drop duplicates while keeping rows with a specific column value, say "something," you can achieve this by first filtering out the rows with that specific value, then dropping duplicates on the rest of the data, and finally merging the results back together. Here’s how you can do it with Pandas:

Bulk dispense duplicates are valid and must remain.


```python
#Filtering out bulk dispenses
keep_rows = fullpharm[fullpharm['Hospital Number'] == 'SH21/0000']
keep_rows
```

```python
fullpharm.drop_duplicates(inplace = True)
```

```python
# Concatenate the rows to keep with the deduplicated rows
result = pd.concat([keep_rows, fullpharm]).reset_index(drop=True)
result
```

```python
result.duplicated().sum()
```
